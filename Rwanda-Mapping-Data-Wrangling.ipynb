### Feature Engineering
> Translating survey responses from multiple choice options into a 1/0 binary for a risk to that child of having a developmental delay

```python
#make all strings lowercase
df = df.applymap(lambda x:x.lower() if type(x) == str else x)
```

```python
#Parsing date of birth and calculating age when data was collected

df['dob']  = pd.to_datetime(df['dob'],errors='coerce')
df['date'] = pd.to_datetime('01/11/2019')
df['age'] = ((df['date']-df['dob']).dt.days)/365

def age(x):
    if x>=100:
        return np.nan
    elif x<=0:
        return np.nan
    else:
        return x

df['age']=df['age'].apply(age)
df['age'].describe()
```

```python
#Recode gender M/F as 1/0
def gender01(x):
    if x=='m':
        return 1
    elif x=='f':
        return 0
    else:
        return np.nan

df['gender']=df['gender'].apply(gender01)
print(df['gender'].value_counts())
```

```python
#Question 11 - Did you/the mother have any pregnancy complications?
def risk_pregcomplication(x):
    if x=='malaria when pregnant':
        return 1
    elif x=='maternal under-nourishment':
        return 1
    elif x=="other illness / infection / undiagnosed problem (describe)":
        return 1
    elif x=='bleeding when pregnant':
        return 1
    elif x=='other illness undiagnosed problem describe':
        return 1
    elif x=='gestational diabetes':
        return 1
    elif x=='hypertension':
        return 1
    elif x=='intra-uterine growth restriction':
        return 1
    elif x=="pre-eclampsia / eclampsia (with or without convulsions)" :
        return 1
    elif x=='pre-eclampsia':
        return 1
    elif x=='meningitis when pregnant':
        return 1
    elif x=='intro-uterine growth restriction':
        return 1
    elif x=='do not know':
        return 1
    elif x=='gestional diabetes':
        return 1
    elif x=="prolonged labour (18 hours+)":
        return 1
    elif x=='hiv positive during pregnancy':
        return 1
    elif x=='other':
        return 1
    elif x=='no known issues':
        return 0
    else:
        return np.nan

df['risk_pregcomplication']=df['pregcomplication_which'].apply(risk_pregcomplication)

print(df['risk_pregcomplication'].value_counts())
```
